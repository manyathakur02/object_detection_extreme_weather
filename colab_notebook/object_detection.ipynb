```python
#training model
from ultralytics import YOLO

# Load pretrained YOLOv8 model
model = YOLO('yolov5s.pt')  # you can change this to yolov5s.pt or yolov8m.pt etc.

# Train your model-9*.-8+
model.train(data='/content/drive/MyDrive/DL_projects/weatherdataset/data.yaml', epochs=50, imgsz=640, name='weather_yolo_model', project='/content/drive/MyDrive/DL_projects/weatherdataset/weather_project')
```

```python
from IPython.display import Image, display
import glob

result_images = glob.glob('/content/drive/MyDrive/DL_projects/weatherdataset/test/images/dusttornado-001_jpg.rf.9d7ff8e49da4685454f963cde9b3ca2e.jpg')
for img_path in result_images[:5]:  # Show first 5
    display(Image(filename=img_path))
```

```python
# prompt: test model on above image path

# You can also test the model directly on an image path using the loaded model object
image_to_test = '/content/drive/MyDrive/DL_projects/weatherdataset/test/images/dusttornado-016_jpg.rf.681cc79480d00f893670ea18dea6d36f.jpg'

# Use the model object for prediction
# The results object contains the detection outputs
results = model(image_to_test)

# You can print or access the results data
for r in results:
    print(r.boxes)  # Print bounding boxes
    print(r.masks) # Print masks if available
    print(r.probs) # Print probabilities if available

# Optionally, save the result image
# results.save_object_predictions() # This is not a standard method, check YOLOv8 docs for saving

# YOLOv8 results objects have a .save() method to save visualized results
results[0].save(filename='detected_dusttornado.jpg')
display(Image(filename='detected_dusttornado.jpg'))
```
